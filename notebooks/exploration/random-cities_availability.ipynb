```python
import geopandas as gpd
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import os
import sys
module_path = os.path.abspath(os.path.join(accesibilidad_urbana))
if module_path not in sys.path:
    sys.path.append(module_path)
    import aup
```

```python
# Load cities (municipalities)
metro_schema = 'metropolis'
metro_table = 'metro_gdf_2020'
metro_query = f"SELECT city FROM {metro_schema}.{metro_table}"
metro_gdf = aup.df_from_query(metro_query)
# Format
metro_gdf.drop_duplicates(inplace=True)
metro_gdf.reset_index(inplace=True,drop=True)
# Show
print(metro_gdf.shape)
print(len(list(metro_gdf.city.unique())))
metro_gdf.head(10)
```

```python
# Load hexs 2010
hex_schema = 'hexgrid'
hex_table = 'hexgrid_8_city'
hex_query = f"SELECT metropolis FROM {hex_schema}.{hex_table}"
hex_df = aup.df_from_query(hex_query)
# Format
hex_df.drop_duplicates(inplace=True)
hex_df.reset_index(inplace=True,drop=True)
# Register
for city in list(hex_df.metropolis.unique()):
    if city in list(metro_gdf.city.unique()):
        idx = metro_gdf.city==city
        metro_gdf.loc[idx,'hex8_2010'] = 1
    else:
        print(f"City {city} not in metro_gdf_2020.")

# Show
print(hex_df.shape)
hex_df.head(2)
```

```python
# Load hexs 2020
hex_schema = 'hexgrid'
hex_table = 'hexgrid_8_city_2020'
hex_query = f"SELECT city FROM {hex_schema}.{hex_table}"
hex_df = aup.df_from_query(hex_query)
# Format
hex_df.drop_duplicates(inplace=True)
hex_df.reset_index(inplace=True,drop=True)
# Register
for city in list(hex_df.city.unique()):
    if city in list(metro_gdf.city.unique()):
        idx = metro_gdf.city==city
        metro_gdf.loc[idx,'hex8_2020'] = 1
    else:
        print(f"City {city} not in metro_gdf_2020.")
# Show
print(hex_df.shape)
hex_df.head(2)
```
