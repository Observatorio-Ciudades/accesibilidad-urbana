```python
import geopandas as gpd
import pandas as pd
```

```python
kmhr = 3.5
m_m = kmhr * 16.666
city = 'Guadalajara'
```

```python
gdf = gpd.read_file(f'../data/processed/{city}_hex_bins.geojson')
```

```python
def get_df(gdf):
    gdf['centroid_coords'] = gdf['geometry'].centroid
    gdf['lat'] = gdf.centroid_coords.apply(lambda x: x.y)
    gdf['lon'] = gdf.centroid_coords.apply(lambda x: x.x)
    gdf['dist_promedio'] = (gdf['dist_supermercados'] + gdf['dist_hospitales'] + gdf['dist_farmacias']) / 3
    gdf['min_farmacias'] = gdf['dist_farmacias']/m_m
    gdf['min_supermercados'] = gdf['dist_supermercados']/m_m
    gdf['min_hospitales'] = gdf['dist_hospitales']/m_m
    gdf['tiempo_promedio'] = (gdf['min_hospitales']+gdf['min_supermercados']+gdf['min_farmacias'])/3
    return pd.merge(gdf.iloc[:,0:6],gdf.iloc[:,-8:-1],left_index=True,right_index=True)
```

```python
df = get_df(gdf)
df
```

```python
df.to_csv(f'../data/csv_web/{city}_data.csv')
```
