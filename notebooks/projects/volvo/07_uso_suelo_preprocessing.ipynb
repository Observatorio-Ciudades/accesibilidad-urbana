Data classification

```python
gdf_gdl.LAYER.unique()
```

```python
def clasificar_layer(uso):
    
    clasificacion = {
    "Residencial": [
        'Habitacional'
    ],
    "ComercialServicios": [
        'Comercio y Servicios',
    ],
    "Equipamiento": [
        'Equipamiento'
    ],
    "Industrial": [
        'Industrial'
    ],
    "Complementarios": [
        'Agropecuario',
        'Espacios Abiertos',
        'Baldios',
        'Infraestructura',
        'Sin actividad registrada'  
    ]
    }
    
    for key, value in clasificacion.items(): 
        for v in value:
            if uso == v: 
                return key
```

```python
gdf_gdl['clasificacion_uso_suelo'] = ''
```

```python
gdf_gdl.loc[:,'clasificacion_uso_suelo'] = gdf_gdl['LAYER'].apply(clasificar_layer)
```
