# Mapa Interactivo con Capas

```python
m = leafmap.Map()
m.add_gdf(buffer, layer_name="Avenida Alameda", zoom_to_layer=True)
m.add_gdf(carabineros, layer_name="Cuarteles Carabineros", zoom_to_layer=True)
m.add_gdf(bomberos, layer_name="Compañias de Bomberos", zoom_to_layer=True)
m.add_gdf(eje, layer_name="Eje Nueva Alameda", zoom_to_layer=True)
m.add_gdf(educ, layer_name="Establecimientos educacionales", zoom_to_layer=True)
m.add_gdf(salud, layer_name="Establecimientos de salud", zoom_to_layer=True)
m.add_gdf(mzas, layer_name="Manzanas de la Metropolitana de Santiago", zoom_to_layer=True)
m.add_gdf(parques, layer_name="Parques urbanos", zoom_to_layer=True)
m.add_gdf(poly_santiago, layer_name="Polígono de Santiago", zoom_to_layer=True)
m.add_gdf(redvial, layer_name="Red Vial", zoom_to_layer=True)
m.add_gdf(uni_operativas, layer_name="Unidades Operativas", zoom_to_layer=True)
m.add_gdf(suelo_mza, layer_name="Usos de Suelo Manzana", zoom_to_layer=True)
m
```

```python
m = folium.Map(location=[poly_santiago['geometry'].centroid.y.mean(), poly_santiago['geometry'].centroid.x.mean()], zoom_start=14.45)


def add_geojson_layer(gdf, name):
    gjson = gdf.to_crs("EPSG:4326").to_json()
    folium.GeoJson(gjson, name=name).add_to(feature_group)
    folium.LayerControl().add_to(m)


feature_group = folium.FeatureGroup(name='Capas')
add_geojson_layer(buffer, "Buffer")
add_geojson_layer(bomberos, "Bomberos")
add_geojson_layer(carabineros, "Carabineros")
add_geojson_layer(eje, "Eje")
add_geojson_layer(educ, "Educación")
add_geojson_layer(salud, "Salud")
add_geojson_layer(mzas, "Manzanas")
add_geojson_layer(parques, "Parques")
add_geojson_layer(poly_santiago, "Santiago")
add_geojson_layer(redvial, "Red Vial")
add_geojson_layer(uni_operativas, "Unidades Operativas")
add_geojson_layer(suelo_mza, "Suelo Manzana")

feature_group.add_to(m)

m.save("Prueba_Mapa.html")

m
```
